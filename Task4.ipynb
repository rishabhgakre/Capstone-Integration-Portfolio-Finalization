## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/Internship/Task4-Data Storytelling & Statistical Validation/raw_superstore_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

## Data Cleaning



```python
df = df.drop_duplicates()
df.isnull().sum()
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
```

## EDA

```python
sales_trend = df.groupby("Year")["Sales"].sum()

sales_trend.plot(kind="line", marker="o", title="Sales Trend")
plt.show()
```

```python
sns.barplot(data=df, x="Category", y="Sales")
plt.title("Sales by Category")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="Sub-Category", y="Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df, x="Region", y="Sales")
plt.title("Sales by Region")
plt.show()
```

```python
sns.barplot(data=df, x="Segment", y="Sales")
plt.show()
```

```python
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_customers.plot(kind="bar", title="Top Customers")
plt.show()
```

 Insight:
 Technology category has highest sales, indicating strong demand.

## Hypothesis Testing

```python
region1 = df[df["Region"] == "West"]["Sales"]
region2 = df[df["Region"] == "East"]["Sales"]
```

```python
from scipy.stats import ttest_ind

stat, p_value = ttest_ind(region1, region2)
print("P-value:", p_value)
```
